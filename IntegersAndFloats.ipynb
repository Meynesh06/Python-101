Arithmetic operations in Python involve mathematical calculations on numerical values. Python supports various arithmetic operators for performing these calculations. Here are the common arithmetic operators along with code examples:

1. **Addition (`+`)**: Adds two numbers.
```python
a = 5
b = 3
result = a + b
print(result)  # Output: 8
```

2. **Subtraction (`-`)**: Subtracts the second number from the first.
```python
a = 10
b = 4
result = a - b
print(result)  # Output: 6
```

3. **Multiplication (`*`)**: Multiplies two numbers.
```python
a = 6
b = 7
result = a * b
print(result)  # Output: 42
```

4. **Division (`/`)**: Divides the first number by the second, result is a float.
```python
a = 15
b = 3
result = a / b
print(result)  # Output: 5.0
```

5. **Floor Division (`//`)**: Divides the first number by the second and rounds down to the nearest integer.
```python
a = 17
b = 4
result = a // b
print(result)  # Output: 4
```

6. **Modulus (`%`)**: Returns the remainder after division.
```python
a = 20
b = 7
result = a % b
print(result)  # Output: 6
```

7. **Exponentiation (`**`)**: Raises the first number to the power of the second.
```python
a = 2
b = 3
result = a ** b
print(result)  # Output: 8
```

8. **Negative Numbers and Parentheses**:
```python
a = 10
b = -3
result = a + (-b)  # Parentheses used for clarity
print(result)  # Output: 13
```

Remember that the order of operations follows the standard rules of mathematics (PEMDAS/BODMAS), where parentheses are evaluated first, then exponents, then multiplication and division (left to right), and finally addition and subtraction (left to right).

```python
result = 2 + 3 * 4  # 2 + (3 * 4) = 14
print(result)       # Output: 14
```

Python also allows you to use the `+=`, `-=`, `*=`, `/=`, `//=`, and other similar compound assignment operators to perform an operation and assign the result to the same variable:

```python
a = 10
a += 5  # Equivalent to a = a + 5
print(a)  # Output: 15
```

These are the basic arithmetic operations in Python. They are fundamental for mathematical calculations in programming.

 Python provides various built-in methods for working with numeric data types. Here are some commonly used numeric methods with code examples:

1. **`abs()`**: Returns the absolute value of a number.
```python
num = -5
abs_num = abs(num)
print(abs_num)  # Output: 5
```

2. **`round()`**: Rounds a floating-point number to the nearest integer.
```python
pi = 3.14159
rounded_pi = round(pi)
print(rounded_pi)  # Output: 3
```

3. **`max()`** and **`min()`**: Returns the maximum and minimum value from a sequence of numbers.
```python
numbers = [5, 10, 2, 8]
max_num = max(numbers)
min_num = min(numbers)
print(max_num)  # Output: 10
print(min_num)  # Output: 2
```

4. **`sum()`**: Returns the sum of all numbers in a sequence.
```python
numbers = [2, 4, 6, 8]
sum_of_numbers = sum(numbers)
print(sum_of_numbers)  # Output: 20
```

5. **`pow()`**: Raises a number to a specified power.
```python
base = 2
exponent = 3
result = pow(base, exponent)  # Equivalent to base ** exponent
print(result)  # Output: 8
```

6. **`divmod()`**: Returns the quotient and remainder of division.
```python
quotient, remainder = divmod(17, 4)
print(quotient, remainder)  # Output: 4 1
```

7. **`int()`**, **`float()`**, and **`str()`**: Convert between different numeric types and strings.
```python
num = 42
float_num = float(num)
str_num = str(num)
print(float_num)  # Output: 42.0
print(str_num)    # Output: "42"
```

8. **`bin()`**, **`oct()`**, and **`hex()`**: Convert integers to binary, octal, and hexadecimal strings.
```python
num = 255
binary = bin(num)
octal = oct(num)
hexadecimal = hex(num)
print(binary)      # Output: '0b11111111'
print(octal)       # Output: '0o377'
print(hexadecimal)  # Output: '0xff'
```

9. **`isinstance()`**: Checks if a value is an instance of a specified numeric type.
```python
num = 10
is_int = isinstance(num, int)
print(is_int)  # Output: True
```

Bitwise operations in Python allow you to manipulate individual bits within integer values. These operations are often used in low-level programming, dealing with flags, and other scenarios where you need to work at the bit level. Here are the bitwise operators with code examples:

1. **Bitwise AND (`&`)**: Performs a bitwise AND operation between corresponding bits of two integers.
```python
a = 5   # 0101 in binary
b = 3   # 0011 in binary
result = a & b
print(result)  # Output: 1 (0001 in binary)
```

2. **Bitwise OR (`|`)**: Performs a bitwise OR operation between corresponding bits of two integers.
```python
a = 5   # 0101 in binary
b = 3   # 0011 in binary
result = a | b
print(result)  # Output: 7 (0111 in binary)
```

3. **Bitwise XOR (`^`)**: Performs a bitwise XOR (exclusive OR) operation between corresponding bits of two integers.
```python
a = 5   # 0101 in binary
b = 3   # 0011 in binary
result = a ^ b
print(result)  # Output: 6 (0110 in binary)
```

4. **Bitwise NOT (`~`)**: Flips the bits of an integer, effectively reversing its binary representation (note that this includes the sign bit, resulting in a different negative value).
```python
a = 5   # 0101 in binary
result = ~a
print(result)  # Output: -6 (1010 in binary)
```

5. **Left Shift (`<<`)**: Shifts the bits of an integer to the left by a specified number of positions. This is equivalent to multiplying the number by 2 to the power of the shift amount.
```python
a = 5   # 0101 in binary
shifted = a << 2
print(shifted)  # Output: 20 (10100 in binary)
```

6. **Right Shift (`>>`)**: Shifts the bits of an integer to the right by a specified number of positions. This is equivalent to dividing the number by 2 to the power of the shift amount (rounded down).
```python
a = 20   # 10100 in binary
shifted = a >> 2
print(shifted)  # Output: 5 (0101 in binary)
```

Bitwise operations are particularly useful for working with flags or setting specific bits to control various options in a compact manner. Keep in mind that the exact behavior of these operations depends on the underlying representation of integers, and they can vary with signed and unsigned integer representations.

Comparison operations in Python are used to compare two values and determine their relationship. These operations return a boolean value (`True` or `False`) based on whether the comparison is true or false. Here are the comparison operators with code examples:

1. **Equal (`==`)**: Checks if two values are equal.
```python
x = 5
y = 5
result = x == y
print(result)  # Output: True
```

2. **Not Equal (`!=`)**: Checks if two values are not equal.
```python
x = 5
y = 3
result = x != y
print(result)  # Output: True
```

3. **Greater Than (`>`)**: Checks if the left value is greater than the right value.
```python
x = 7
y = 5
result = x > y
print(result)  # Output: True
```

4. **Less Than (`<`)**: Checks if the left value is less than the right value.
```python
x = 3
y = 5
result = x < y
print(result)  # Output: True
```

5. **Greater Than or Equal To (`>=`)**: Checks if the left value is greater than or equal to the right value.
```python
x = 7
y = 5
result = x >= y
print(result)  # Output: True
```

6. **Less Than or Equal To (`<=`)**: Checks if the left value is less than or equal to the right value.
```python
x = 3
y = 5
result = x <= y
print(result)  # Output: True
```

7. **Chaining Comparisons**: You can chain multiple comparisons together using the `and` and `or` operators.
```python
x = 10
y = 5
z = 7
result = x > y and y < z
print(result)  # Output: True
```

8. **`in` and `not in`**: Checks if a value is present (or not present) in a sequence like a list or a string.
```python
my_list = [1, 2, 3, 4, 5]
result1 = 3 in my_list
result2 = 6 not in my_list
print(result1)  # Output: True
print(result2)  # Output: True
```

These comparison operators are fundamental in programming as they allow you to make decisions based on the relationships between different values. They are often used in `if` statements, loops, and other control structures.
